```python
# import necessary library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("---------------------------------------------")

# load the dataset
data = pd.read_csv("/content/SuperMarket Analysis.csv")

print("---------------------------------------------")

# View Basic Info
print(data.head())            # show first 5 rows
print(data.info())            # Shows column names, data types, missing values
print(data.describe())        # Gives basic statistics like mean, min, max, etc.

print("---------------------------------------------")

# Calculate Metrics Using NumPy
mean_total = np.mean(data['Sales'])    # Average sales amount across all transactions
max_rating = np.max(data['Rating'])    # Highest customer rating recorded

print("---------------------------------------------")

print(f"Average Sales Amount: {mean_total}")
print(f"Maximum Customer Rating: {max_rating}")

print("---------------------------------------------")

# Analyze Sales by Gender
sales_by_gender = data.groupby("Gender")['Sales'].sum()  #Groups the data by Gender (e.g., Male/Female).
print(sales_by_gender)                                   #Calculates total sales made by each gender using sum()

print("---------------------------------------------")

# Plot Total Sales by Product Line
product_sales = data.groupby("Product line")["Sales"].sum()  # group by Total sales per product line

plt.figure(figsize=(8,5))
product_sales.plot(kind="bar", color="skyblue")
plt.title("Total Sales by Product Line")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)                 # Rotate x-axis labels by 45
plt.show()

print("---------------------------------------------")

# Pie Chart of Payment Methods
payment_counts = data["Payment"].value_counts()     # Count how many times each payment method is used

plt.figure(figsize=(6,6))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%')     # Pie chart with % labels
plt.title("Payment Method Distribution")
plt.show()


# Conclusion
# The project showed how using simple data analysis can help us understand the business better and make smart decisions based on the data.
```
